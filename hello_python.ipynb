```python
#사칙연산 클래스

class Calculator():
  def __init__(self,first,second):
    self.first=first
    self.second=second

#숫자 두 개 입력 받음 set_data()
  def set_data(self,first,second):
    self.first=first
    self.second=second

#덧셈 plus()
  def plus(self):
    result=self.first+self.second
    return result

#뺄셈 minus()
  def minus(self):
    result=self.first-self.second
    return result

#곱셈 mul()
  def mul(self):
    result=self.first*self.second
    return result

#나눗셈 div()
  def div(self):
    result=self.first/self.second
    return result

calc1=Calculator(5,5)
calc1.set_data(1,2)
calc1.plus()
calc1.minus()
calc1.mul()


print(calc1.first)
print(calc1.second)

calc2=Calculator(3,5)
calc2.set_data(3,5)

print(calc2.first)
print(calc2.second)
print(calc2.mul())

calc3=Calculator(2,3)
calc3.set_data(2,3)
calc3.plus()

class More_calc(Calculator):
  #제곱 연산
  def pow(self):
    result=self.first ** self.second
    return result

class safe_calc(Calculator):
  def div(self):
    if self.second==0:
      return 0
    else:
      result=self.first / self.second
      return result

more_cal1=More_calc(2,4)
print(more_cal1.first)
print(more_cal1.second)
print(more_cal1.mul())
print(more_cal1.div())
print(more_cal1.pow())
```

```python
#10미만의 자연수에서 3과 5의 배수를 구하면 3,5,6,9이다
#이들의 총합은 23이다
#1000 미만의 자연수에서 3의 배수와 5의 배수의 총합을 구하라

def gugu():
  sum=0
  for i in range(1,1000):
    if (i%3==0 or i%5==0):
      sum+=i
  return sum

print(gugu())
```

```python
#게시판 페이징하기
#출력 프로그램
#게시물의 총 개수
#한 페이지 수에 보여 줄 게시물 수를 입력

#총 페이지 수 출력 프로그램
#함수 이름
#입력 값 : 총 개수 / 페이지 당 개수
#출력 값 : 총 페이지 수

def total_pages(sum,n):
  if sum%n==0:
    return sum//n
  else:
    return (sum//n)+1

total_pages(30,10)
# 3
total_pages(5,10)
# 1
```

```python
#함수의 실행 시간 측정하기
import time

def elapsed(org_func):
  def wrapper():
    #시작 시각
    start_time=time.time()
    result=org_func()
    #종료 시각
    end_time=time.time()

    print(f"함수 수행 시간:{end_time - start_time}초입니다.")

    return result
  return wrapper

def my_func():
  print("함수가 실행됩니다")

func_decorated = elapsed(my_func)

func_decorated()
```

```python
# @사용하기
import time

def elapsed(org_func):
  def wrapper():
    #시작 시각
    start_time=time.time()
    result=org_func()
    #종료 시각
    end_time=time.time()

    print(f"함수 수행 시간:{end_time - start_time}초입니다.")

    return result
  return wrapper

# 사용법 @+함수명
@elapsed
def my_func():
  print("함수가 실행됩니다")

# 호출하기
my_func()
```

```python
# @사용하기
import time

def elapsed(org_func):
  def wrapper(*argc,**kwargs):
    #시작 시각
    start_time=time.time()
    result=org_func(*argc,**kwargs)
    #종료 시각
    end_time=time.time()

    print(f"함수 수행 시간:{end_time - start_time}초입니다.")

    return result
  return wrapper

# 사용법 @+함수명
@elapsed
def my_func(m):
  print(f"{m}을 출력합니다")

# 호출하기
my_func("You need python")

@elapsed
def my_func2():
  print("출력합니다")

# 호출하기
my_func2()
```

```python
class Myiterator:
  def __init__(self, data):
    self.data=data
    self.position=0

  def __iter__(self):
    return self

  def __next__(self):
    if self.position >=len(self.data):
      raise StopIteration
    result=self.data[self.position]
    self.position+=1
    return result

i = Myiterator([1,2,3])
for item in i:
   print(item)
```

```python
class Reverseiterator:
  def __init__(self, data):
    self.data=data
    self.position=len(self.data)-1

  def __iter__(self):
    return self

  def __next__(self):
    if self.position <0:
      raise StopIteration
    result=self.data[self.position]
    self.position-=1
    return result

i = Reverseiterator([1,2,3])
for item in i:
   print(item)
```

###정규표현식

```python
#정규 표현식은 왜 필요한가?
#주민등록번호를 포함하고 있는 텍스트가 있다
#이 텍스트에 포함된 모든 주민등록번호의 뒷자리를 *문자로 변경해보자

data="""
park 800953-1049118
kim 700953-1029442
lee 000103-3040505
"""

#전체 텍스트를 공백 문자로 나눈다
result=[]
for line in data.split("\n") :
  #print(line)
  result_word=[]
  for word in line.split(" "):
    #나눈 단어가 주민등록번호 형식인가 확인
    if len(word)==14 and word[:6].isdigit() and word[7:].isdigit() :
      #if 단어가 주민등록버호 형식이면 뒷자리를 *로 변환한다. (replace)
      word=word[:6]+"-" "*******"
      #print(word)
    result_word.append(word)
  print(result_word)
  #나뉜 단어를 다시 조립한다
  result.append(" ".join(result_word))
print("\n".join(result))
```

```python
# -*- coding: utf-8 -*-
"""
Created on Tue Feb 13 11:34:35 2024

@author: microsoft
"""

import numpy as np
import tensorflow as tf
import tensorflow.keras.datasets as ds

from tensorflow.keras.models import Sequential
# 컨볼루션층, 풀링층, Flattern 연산에 사용할 클래스
from tensorflow.keras.layers import Conv2D,MaxPooling2D,Flatten,Dropout,Dense
from tensorflow.keras.optimizers import Adam

(x_train,y_train),(x_test,y_test)=ds.mnist.load_data()
x_train=x_train.reshape(60000,28,28,1)
x_test=x_test.reshape(10000,28,28,1) #3차원 구조로 변환
x_train=x_train.astype(np.float32)/255.0
x_test=x_test.astype(np.float32)/255.0
y_train=tf.keras.utils.to_categorical(y_train,10)
y_test=tf.keras.utils.to_categorical(y_test,10)

#LeNet 신경망 구조를 만든다
cnn=Sequential() #컨볼루션층(5*5 커널을 6개 사용)
cnn.add(Conv2D(6,(5,5),padding='same',activation='relu',input_shape=(28,28,1)))

#padding='same' 0 패딩을 적용
#input_shape = (28,28,1) 최초 입력되는 텐서 모양 지
cnn.add(MaxPooling2D(pool_size=(2,2),strides=2))
cnn.add(Conv2D(16,(5,5),padding='valid',activation='relu'))
# padding='valid'덧대기를 적용하지 말아라
cnn.add(MaxPooling2D(pool_size=(2,2),strides=2))
cnn.add(Conv2D(120,(5,5),padding='valid',activation='relu'))
cnn.add(Flatten()) #1차원 구조로 변환
cnn.add(Dense(units=84,activation='relu'))
cnn.add(Dense(units=10,activation='softmax'))

#학습
cnn.compile(loss='categorical_crossentropy',optimizer=Adam(learning_rate=0.001),metrics=['accuracy'])
cnn.fit(x_train,y_train,batch_size=128,epochs=30,validation_data=(x_test,y_test),verbose=2)

#예측(성능 측정)
res=cnn.evaluate(x_test,y_test,verbose=0)
print('정확률=',res[1]*100)
#정확률= 99.04999732971191
```
